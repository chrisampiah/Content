Convolution

Have a look at the following [tutorial](https://) and use convolution to modify your image.


```python
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt
import cv2

# Load image
image_path = '/content/drive/MyDrive/Thesis_Week1/image_array.jpg'
img = Image.open(image_path).convert('L')  # convert to grayscale
img = np.array(img)

# Define a kernel (example: edge detection)
kernel = np.array([[ -1, -1, -1],
                   [ -1,  8, -1],
                   [ -1, -1, -1]])

# Apply convolution using OpenCV
convolved = cv2.filter2D(img, -1, kernel)

# Show results
plt.figure(figsize=(10,5))

plt.subplot(1,2,1)
plt.title("Original Image")
plt.imshow(img, cmap='gray')
plt.axis('off')

plt.subplot(1,2,2)
plt.title("After Convolution (Edge Detection)")
plt.imshow(convolved, cmap='gray')
plt.axis('off')

plt.show()
```
