```python
# ============================================================
# DATA DOWNLOAD + SAFE EXTRACTION (ROBUST VERSION)
# ============================================================

import os
import zipfile

DATA_DIR = "/content/myelin_dataset"
ZIP_NAME = "zenodo_upload_v2.zip"
ZIP_PATH = os.path.join(DATA_DIR, ZIP_NAME)
EXTRACT_FLAG = os.path.join(DATA_DIR, ".extracted")

os.makedirs(DATA_DIR, exist_ok=True)

# ------------------------------------------------------------
# 1. DOWNLOAD (only if missing)
# ------------------------------------------------------------
if not os.path.exists(ZIP_PATH):
    print("Downloading dataset from Zenodo...")
    !zenodo_get 10.5281/zenodo.8055666 -o {DATA_DIR}
else:
    print("Dataset already downloaded.")

# ------------------------------------------------------------
# 2. VERIFY ZIP (basic integrity check)
# ------------------------------------------------------------
if os.path.exists(ZIP_PATH):
    try:
        with zipfile.ZipFile(ZIP_PATH, 'r') as z:
            bad_file = z.testzip()
            if bad_file:
                raise RuntimeError(f"Corrupt file detected: {bad_file}")
        print("Zip integrity check passed.")
    except Exception as e:
        raise RuntimeError("Zip file is corrupted. Please delete and re-download.") from e
else:
    raise FileNotFoundError("Zip file not found after download!")

# ------------------------------------------------------------
# 3. EXTRACT ONLY ONCE
# ------------------------------------------------------------
if not os.path.exists(EXTRACT_FLAG):
    print("Extracting dataset (this will take a few minutes)...")

    with zipfile.ZipFile(ZIP_PATH, 'r') as zip_ref:
        zip_ref.extractall(DATA_DIR)

    # Create flag file
    with open(EXTRACT_FLAG, "w") as f:
        f.write("done")

    print("Extraction complete.")
else:
    print("Dataset already extracted. Skipping extraction.")

# ------------------------------------------------------------
# 4. VERIFY STRUCTURE
# ------------------------------------------------------------
subjects = [d for d in os.listdir(DATA_DIR) if d.startswith("s")]
print(f"\nTotal subject folders detected: {len(subjects)}")

# Show sample
print("\nSample structure:")
for s in subjects[:5]:
    print(os.path.join(DATA_DIR, s))
```

```python
# ============================================================
# METADATA LOADING + VALIDATION (CLEAN VERSION)
# ============================================================

import pandas as pd
import glob
import os

# ------------------------------------------------------------
# 1. FIND CSV FILE (ROBUST)
# ------------------------------------------------------------
csv_files = glob.glob(f"{DATA_DIR}/**/*.csv", recursive=True)

if len(csv_files) == 0:
    raise FileNotFoundError("No CSV file found in dataset directory!")

# Prefer file containing 'meta' in name (safer)
meta_candidates = [f for f in csv_files if 'meta' in os.path.basename(f).lower()]

if len(meta_candidates) > 0:
    meta_path = meta_candidates[0]
else:
    meta_path = csv_files[0]

print(f"Using metadata file: {meta_path}")

# ------------------------------------------------------------
# 2. LOAD DATA
# ------------------------------------------------------------
df = pd.read_csv(meta_path, delimiter=';')

# ------------------------------------------------------------
# 3. BASIC INSPECTION
# ------------------------------------------------------------
print("\nShape:", df.shape)
print("\nColumns:", df.columns.tolist())

# ------------------------------------------------------------
# 4. CLEANING (CRITICAL)
# ------------------------------------------------------------
# Strip whitespace from column names
df.columns = df.columns.str.strip()

# Standardize text columns
for col in ['myelinisation', 'diagnosis', 'group']:
    if col in df.columns:
        df[col] = df[col].astype(str).str.strip().str.lower()

# ------------------------------------------------------------
# 5. VALIDATE REQUIRED COLUMNS
# ------------------------------------------------------------
required_cols = ['image_id', 'myelinisation']

missing = [c for c in required_cols if c not in df.columns]
if missing:
    raise ValueError(f"Missing required columns: {missing}")

# ------------------------------------------------------------
# 6. LABEL DISTRIBUTION (VERY IMPORTANT)
# ------------------------------------------------------------
print("\nMyelinisation distribution:")
print(df['myelinisation'].value_counts())

# ------------------------------------------------------------
# 7. CHECK FOR MISSING VALUES
# ------------------------------------------------------------
print("\nMissing values per column:")
print(df.isnull().sum())

# ------------------------------------------------------------
# 8. PREVIEW
# ------------------------------------------------------------
df.head()
```

```python
# ============================================================
# BLOCK 4: LABEL CREATION + PATH BUILDING (FINAL CLEAN VERSION)
# ============================================================

import os
import numpy as np
import pandas as pd

# ------------------------------------------------------------
# 1. USE TRUE LABELS (CRITICAL FIX)
# ------------------------------------------------------------
df_bin = df.copy()

# Keep only valid classes
df_bin = df_bin[df_bin['myelinisation'].isin(['normal', 'delayed'])].copy()

# Create binary label
df_bin['label'] = (df_bin['myelinisation'] == 'delayed').astype(int)

print("\nLabel distribution (TRUE labels):")
print(df_bin['label'].value_counts())

# ------------------------------------------------------------
# 2. OPTIONAL: KEEP AGE FEATURES (DO NOT USE AS LABEL)
# ------------------------------------------------------------
df_bin['age'] = pd.to_numeric(df_bin['age'], errors='coerce')
df_bin['doctor_predicted_age'] = pd.to_numeric(df_bin['doctor_predicted_age'], errors='coerce')

df_bin['age_gap'] = df_bin['age'] - df_bin['doctor_predicted_age']

# ------------------------------------------------------------
# 3. BUILD FILE PATHS (FAST + RELIABLE)
# ------------------------------------------------------------
t1_paths, t2_paths, valid_idx = [], [], []

for i, row in df_bin.iterrows():
    sid = str(row['image_id']).strip()

    t1_path = os.path.join(DATA_DIR, sid, "t1.nii.gz")
    t2_path = os.path.join(DATA_DIR, sid, "t2.nii.gz")

    if os.path.exists(t1_path) and os.path.exists(t2_path):
        t1_paths.append(t1_path)
        t2_paths.append(t2_path)
        valid_idx.append(i)
    else:
        # Optional debug:
        # print(f"Missing files for subject {sid}")
        pass

# Filter valid subjects
df_valid = df_bin.loc[valid_idx].copy()
df_valid['t1_path'] = t1_paths
df_valid['t2_path'] = t2_paths

print(f"\nSubjects with both T1w + T2w: {len(df_valid)}")

# ------------------------------------------------------------
# 4. FINAL SANITY CHECKS (IMPORTANT)
# ------------------------------------------------------------
print("\nFinal label distribution (after filtering):")
print(df_valid['label'].value_counts())

# Check for missing paths
assert df_valid['t1_path'].isnull().sum() == 0, "Missing T1 paths!"
assert df_valid['t2_path'].isnull().sum() == 0, "Missing T2 paths!"

# Preview
df_valid.head()
```

```python
# ============================================================
# BLOCK 5: MONAI TRANSFORMS (PUBLICATION-LEVEL VERSION)
# ============================================================

from monai.transforms import (
    Compose, LoadImaged, EnsureChannelFirstd,
    Spacingd, Orientationd,
    ScaleIntensityRangePercentilesd,
    Resized, ConcatItemsd, ToTensord,

    # AUGMENTATION (IMPORTANT)
    RandFlipd, RandAffined, RandGaussianNoised,
    RandBiasFieldd
)

TARGET_SHAPE = (96, 96, 96)  # increased resolution (better anatomical detail)

# ------------------------------------------------------------
# TRAIN TRANSFORMS (WITH AUGMENTATION)
# ------------------------------------------------------------
train_transforms = Compose([
    LoadImaged(keys=['t1', 't2']),
    EnsureChannelFirstd(keys=['t1', 't2']),
    Orientationd(keys=['t1', 't2'], axcodes='RAS'),

    Spacingd(keys=['t1', 't2'], pixdim=(2.0, 2.0, 2.0), mode=('bilinear', 'bilinear')),

    ScaleIntensityRangePercentilesd(
        keys=['t1', 't2'],
        lower=1, upper=99,
        b_min=0.0, b_max=1.0,
        clip=True
    ),

    # ---------------- AUGMENTATION ----------------
    RandFlipd(keys=['t1', 't2'], prob=0.5, spatial_axis=0),
    RandFlipd(keys=['t1', 't2'], prob=0.5, spatial_axis=1),

    RandAffined(
        keys=['t1', 't2'],
        prob=0.3,
        rotate_range=(0.1, 0.1, 0.1),
        scale_range=(0.1, 0.1, 0.1),
        mode=('bilinear', 'bilinear')
    ),

    RandGaussianNoised(keys=['t1', 't2'], prob=0.2, mean=0.0, std=0.01),

    RandBiasFieldd(keys=['t1', 't2'], prob=0.2, coeff_range=(0.0, 0.5)),

    # ---------------- RESIZE + FORMAT ----------------
    Resized(keys=['t1', 't2'], spatial_size=TARGET_SHAPE),

    ConcatItemsd(keys=['t1', 't2'], name='image', dim=0),

    ToTensord(keys=['image'])
])

# ------------------------------------------------------------
# VALIDATION TRANSFORMS (NO AUGMENTATION)
# ------------------------------------------------------------
val_transforms = Compose([
    LoadImaged(keys=['t1', 't2']),
    EnsureChannelFirstd(keys=['t1', 't2']),
    Orientationd(keys=['t1', 't2'], axcodes='RAS'),

    Spacingd(keys=['t1', 't2'], pixdim=(2.0, 2.0, 2.0), mode=('bilinear', 'bilinear')),

    ScaleIntensityRangePercentilesd(
        keys=['t1', 't2'],
        lower=1, upper=99,
        b_min=0.0, b_max=1.0,
        clip=True
    ),

    Resized(keys=['t1', 't2'], spatial_size=TARGET_SHAPE),

    ConcatItemsd(keys=['t1', 't2'], name='image', dim=0),

    ToTensord(keys=['image'])
])
```

```python
# ============================================================
# BLOCK 6: DATASET + SPLITS + DATALOADERS (IMPROVED VERSION)
# ============================================================

import torch
import numpy as np
from torch.utils.data import Dataset, DataLoader, WeightedRandomSampler
from sklearn.model_selection import train_test_split

# ------------------------------------------------------------
# 1. REPRODUCIBILITY (IMPORTANT FOR THESIS)
# ------------------------------------------------------------
seed = 42
torch.manual_seed(seed)
np.random.seed(seed)

# ------------------------------------------------------------
# 2. DATASET CLASS (UNCHANGED BUT CLEAN)
# ------------------------------------------------------------
class MyelinDataset(Dataset):
    def __init__(self, dataframe, transforms):
        self.df = dataframe.reset_index(drop=True)
        self.transforms = transforms

    def __len__(self):
        return len(self.df)

    def __getitem__(self, idx):
        row = self.df.iloc[idx]

        data = {
            't1': row['t1_path'],
            't2': row['t2_path']
        }

        data = self.transforms(data)

        image = data['image']
        label = torch.tensor(row['label'], dtype=torch.long)

        return image, label

# ------------------------------------------------------------
# 3. SAFETY CHECK
# ------------------------------------------------------------
print("\nLabel distribution:")
print(df_valid['label'].value_counts())

if df_valid['label'].nunique() < 2:
    raise ValueError("Only one class found!")

# ------------------------------------------------------------
# 4. STRATIFIED SPLIT (IMPROVED BALANCE)
# ------------------------------------------------------------
train_df, temp_df = train_test_split(
    df_valid,
    test_size=0.50,   # Changed from 0.30 to 0.50 to ensure enough minority samples in temp_df
    stratify=df_valid['label'],
    random_state=seed
)

val_df, test_df = train_test_split(
    temp_df,
    test_size=0.50,
    stratify=temp_df['label'],
    random_state=seed
)

print(f"\nTrain: {len(train_df)} | Val: {len(val_df)} | Test: {len(test_df)}")

# ------------------------------------------------------------
# 5. STRONG CLASS IMBALANCE HANDLING
# ------------------------------------------------------------

# Class counts
counts = train_df['label'].value_counts().sort_index().values

# LOSS WEIGHTS (keep but stabilize)
class_weights = torch.tensor(
    (1.0 / counts) * (len(counts) / np.sum(1.0 / counts)),
    dtype=torch.float32
) # Removed .cuda() as PyTorch is CPU-only

print("\nClass weights:", class_weights)

# ------------------------------------------------------------
# 6. WEIGHTED SAMPLER (VERY IMPORTANT FIX)
# ------------------------------------------------------------
sample_weights = train_df['label'].map(
    lambda x: 1.0 / counts[x]
).values

sampler = WeightedRandomSampler(
    weights=sample_weights,
    num_samples=len(sample_weights),
    replacement=True
)

# ------------------------------------------------------------
# 7. DATASETS
# ------------------------------------------------------------
BATCH = 2  # safer for 3D MRI

train_ds = MyelinDataset(train_df, train_transforms)
val_ds   = MyelinDataset(val_df, val_transforms)
test_ds  = MyelinDataset(test_df, val_transforms)

# ------------------------------------------------------------
# 8. DATALOADERS
# ------------------------------------------------------------

train_loader = DataLoader(
    train_ds,
    batch_size=BATCH,
    sampler=sampler,   # 🔥 KEY FIX (replaces shuffle)
    num_workers=2,
    pin_memory=True
)

val_loader = DataLoader(
    val_ds,
    batch_size=BATCH,
    shuffle=False,
    num_workers=2,
    pin_memory=True
)

test_loader = DataLoader(
    test_ds,
    batch_size=BATCH,
    shuffle=False,
    num_workers=2,
    pin_memory=True
)

# ------------------------------------------------------------
# 9. SANITY CHECK
# ------------------------------------------------------------
images, labels = next(iter(train_loader))

print("\nSanity check:")
print("Image shape:", images.shape)  # (B, 2, D, H, W)
print("Labels:", labels[:10])
```

```python
from monai.networks.nets import DenseNet121
import torch

device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')

model = DenseNet121(
    spatial_dims=3,
    in_channels=2,
    out_channels=2
).to(device)

print(f"Using DenseNet121 | Params: {sum(p.numel() for p in model.parameters()):,}")
```

```python
# ============================================================
# BLOCK 7: LOSS + OPTIMIZER (IMPROVED FOR IMBALANCE + ROC)
# ============================================================

import torch
from torch.optim import AdamW
from torch.optim.lr_scheduler import CosineAnnealingWarmRestarts

# --------------------------------------------------
# 1. STRONG CLASS WEIGHTING (NO NORMALIZATION)
# --------------------------------------------------
counts = df_valid['label'].value_counts().sort_index().values

class_weights = torch.tensor(
    1.0 / counts,
    dtype=torch.float32
).to(device)

print("\nClass weights (RAW):", class_weights)

# --------------------------------------------------
# 2. LABEL SMOOTHED LOSS (STABILIZES ROC)
# --------------------------------------------------
criterion = torch.nn.CrossEntropyLoss(
    weight=class_weights,
    label_smoothing=0.1   # 🔥 important fix
)

# --------------------------------------------------
# 3. OPTIMIZER (GOOD DEFAULT FOR MEDICAL IMAGING)
# --------------------------------------------------
optimizer = AdamW(
    model.parameters(),
    lr=3e-4,              # slightly higher = better escape from bad minima
    weight_decay=1e-4
)

# --------------------------------------------------
# 4. IMPROVED SCHEDULER (MORE ROBUST THAN COSINE ONLY)
# --------------------------------------------------
scheduler = CosineAnnealingWarmRestarts(
    optimizer,
    T_0=10,   # restart every 10 epochs
    T_mult=1,
    eta_min=1e-6
)
```

```python
# ============================================================
# BLOCK 8: TRAINING LOOP (PUBLICATION-LEVEL STABLE VERSION)
# ============================================================

from sklearn.metrics import balanced_accuracy_score, roc_auc_score, roc_curve
import torch.nn.functional as F
import numpy as np
import torch

EPOCHS = 30
SAVE_PATH = "/content/best_myelin_model.pth"

best_val_auc = 0.0

history = {
    'train_loss': [],
    'val_loss': [],
    'val_bacc': [],
    'val_auc': []
}

# ------------------------------------------------------------
# FIXED EPOCH RUN FUNCTION
# ------------------------------------------------------------
def run_epoch(loader, training=True):

    model.train() if training else model.eval()

    total_loss = 0
    all_labels = []
    all_probs = []

    for images, labels in loader:
        images = images.to(device)
        labels = labels.to(device)

        with torch.set_grad_enabled(training):
            logits = model(images)
            loss = criterion(logits, labels)

            if training:
                optimizer.zero_grad()
                loss.backward()
                torch.nn.utils.clip_grad_norm_(model.parameters(), 1.0)
                optimizer.step()

        total_loss += loss.item() * labels.size(0)

        # --------------------------------------------------------
        # STABLE PROBABILITY EXTRACTION
        # --------------------------------------------------------
        probs = F.softmax(logits, dim=1)[:, 1]

        # 🔥 Clip probabilities to avoid numerical instability
        probs = torch.clamp(probs, 1e-6, 1 - 1e-6)

        all_probs.extend(probs.detach().cpu().numpy())
        all_labels.extend(labels.cpu().numpy())

    avg_loss = total_loss / len(loader.dataset)

    all_labels = np.array(all_labels)
    all_probs = np.array(all_probs)

    # ------------------------------------------------------------
    # SAFE AUC (handle single-class edge cases)
    # ------------------------------------------------------------
    auc = roc_auc_score(all_labels, all_probs) if len(np.unique(all_labels)) > 1 else 0.0

    # ------------------------------------------------------------
    # FIXED: DO NOT OPTIMIZE THRESHOLD EVERY EPOCH
    # ------------------------------------------------------------
    fixed_threshold = 0.5
    preds = (all_probs >= fixed_threshold).astype(int)

    bacc = balanced_accuracy_score(all_labels, preds)

    return avg_loss, bacc, auc


# ============================================================
# TRAINING LOOP
# ============================================================
for epoch in range(1, EPOCHS + 1):

    tr_loss, tr_bacc, tr_auc = run_epoch(train_loader, training=True)
    vl_loss, vl_bacc, vl_auc = run_epoch(val_loader, training=False)

    scheduler.step()

    history['train_loss'].append(tr_loss)
    history['val_loss'].append(vl_loss)
    history['val_bacc'].append(vl_bacc)
    history['val_auc'].append(vl_auc)

    # ------------------------------------------------------------
    # MODEL SELECTION (AUC ONLY - CLEAN AND STANDARD)
    # ------------------------------------------------------------
    if vl_auc > best_val_auc:
        best_val_auc = vl_auc
        torch.save(model.state_dict(), SAVE_PATH)
        flag = " ✓ saved"
    else:
        flag = ""

    print(
        f"Epoch {epoch:02d}/{EPOCHS} | "
        f"Train loss {tr_loss:.4f} | "
        f"Val loss {vl_loss:.4f} | "
        f"Bal-Acc {vl_bacc:.3f} | "
        f"AUC {vl_auc:.3f}{flag}"
    )
```
